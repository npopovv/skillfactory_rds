# 3. Анализ номинативных переменных

Все остальные признаки - категориальные. Для их рассмотрения используем box-plot и t-test.

```python
def get_boxplot(column):
    fig, ax = plt.subplots(figsize=(14, 4))
    sns.boxplot(x=column, y='score',
                data=data, ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for ' + column)
    plt.show()
```

```python
data
```

```python
# Дозаполним все пропущенные значения в колонках модами
for column in data.columns:
    data[str(column)] = data[str(column)].apply(
        lambda x: data[str(column)].mode()[0] if pd.isna(x) else x)
```

```python
data
```

```python
for col in ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu', 'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'studytime', 'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic', 'famrel', 'freetime', 'goout', 'health']:
    get_boxplot(col)
```

Видно, что некоторые признаки влияют на score, некоторые - нет. Однозначно сказать нельзя, воспользуемся t-testом.

Нулевая гипотеза - распределение оценок по различным параметрам неразличимы.

```python
from itertools import combinations
from scipy.stats import ttest_ind
```

```python
def get_stat_dif(column):
    cols = data.loc[:, column].value_counts().index[:]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(data.loc[data.loc[:, column] == comb[0], 'score'],
                     data.loc[data.loc[:, column] == comb[1], 'score']).pvalue \
                <= 0.05/len(combinations_all):
            print('Найдены статистически значимые различия для колонки', column)
            break
```

```python
for col in ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu', 'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'studytime', 'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic', 'famrel', 'freetime', 'goout', 'health']:
    get_stat_dif(col)
```

Итак, серьезно отличаются параметры: sex, address, Medu, Fedu, Mjob, Fjob, studytime, failures, schoolsup, goout. Оставим эти переменные в датасете для дальнейшего построения модели.
